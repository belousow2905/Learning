<br><br>
&nbsp;&nbsp;&nbsp;&nbsp; Любой бизнес хочет максимизировать количество клиентов. Для достижения этой цели важно не только пытаться привлечь новых, но и удерживать уже существующих. Удержать клиента обойдется компании дешевле, чем привлечь нового. Кроме того, новый клиент может оказаться слабо заинтересованным в услугах бизнеса и с ним будет сложно работать, тогда как о старых клиентах уже есть необходимые данные по взаимодействию с сервисом. 
 <br>

&nbsp;&nbsp;&nbsp;&nbsp; Соответственно, прогнозируя отток, мы можем вовремя среагировать и попытаться удержать клиента, который хочет уйти. Опираясь на данные об услугах, которыми пользуется клиент, мы можем сделать ему специальное предложение, пытаясь изменить его решение об уходе от оператора. Благодаря этому задача удержания будет легче в реализации, чем задача привлечения новых пользователей, о которых мы еще ничего не знаем.<br>

&nbsp;&nbsp;&nbsp;&nbsp; Вам предоставлен набор данных от телекоммуникационной компании. В данных содержится информация о почти шести тысячах пользователей, их демографических характеристиках, услугах, которыми они пользуются, длительности пользования услугами оператора, методе оплаты, размере оплаты. 
<br>

&nbsp;&nbsp;&nbsp;&nbsp; Cтоит задача проанализировать данные и спрогнозировать отток пользователей (выявить людей, которые продлят контракт и которые не продлят). Работа должна включать в себя следующие обязательные пункты: 
1. Описание данных *(с расчетом базовых статистик)*,

3. Построение моделей для прогнозирования оттока *(с обоснованием выбора той или иной модели)* на основе проверенных гипотез и выявленных взаимосвязей, 
4. Сравнение качества полученных моделей. 

Во втором разделе обязательно должно присутствовать обоснование гипотез, подробное описание выявленных взаимосвязей, а также их визуализация.

# 3. Построение моделей для прогнозирования оттока

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop('Churn', axis=1)
y=data.Churn
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y,random_state=17)
print(f'X_train shape: {x_train.shape}, y_train shape: {y_train.shape},\n'
      f'X_test shape: {x_test.shape}, y_test shape: {y_test.shape}')
```

В качестве алгоритмов используем RandomForestClassifier, SVC, GradientBoostingClassifier, LogisticRegression и KNeighborsClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

Обучим алгоритмы, для подбора гиперпараметров модели используем метод GridSearchCV() и выведим лучшее сочетание параметров с помощью метода best_params_. Выведим отдельно метрику "доля правильных ответов" Accuracy для каждого алгоритма.

```python
from sklearn.model_selection import GridSearchCV
gbc_params = {'learning_rate': np.arange(0.1, 0.6, 0.1),  # GradientBoostingClassifier
                'max_depth' : np.arange(1, 4, 1) }
```

```python
gbc = GradientBoostingClassifier(random_state=17, n_estimators=100)
```

```python
gbc_grid = GridSearchCV(gbc, gbc_params, cv=5)
```

```python
model_gbc=gbc_grid.fit(x_train, y_train)
```

```python
model_gbc.best_params_
```

```python
predict_gbc = model_gbc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(f'GradientBoostingClassifier Accuracy: { accuracy_score(y_test, predict_gbc)}')
```

```python
rfc_params = {'n_estimators': range(10, 100, 10), # RandomForestClassifier
              'min_samples_leaf': range(1, 5)}
```

```python
rfc= RandomForestClassifier(random_state=17)
```

```python
rfc_grid = GridSearchCV(rfc, rfc_params, cv=5)
```

```python
model_rfc=rfc_grid.fit(x_train, y_train)
```

```python
model_rfc.best_params_
```

```python
predict_rfc = model_rfc.predict(x_test)
```

```python
print(f'RandomForestClassifier Accuracy: { accuracy_score(y_test, predict_rfc)}')
```

```python
svc_params = {'C': np.arange(0.1, 1, 0.2)}  # SVC              
```

```python
svc=SVC(random_state=17, gamma='auto', probability=True)
```

```python
svc_grid = GridSearchCV(svc, svc_params, cv=5)
```

```python
model_svc=svc_grid.fit(x_train, y_train)
```

```python
model_svc.best_params_
```

```python
predict_svc = model_svc.predict(x_test)
```

```python
print(f'SVC Accuracy: { accuracy_score(y_test, predict_svc)}')
```

```python
lr_params = {'C': np.arange(0.5, 1, 0.1)}   # LogisticRegression
```

```python
lr=LogisticRegression(random_state=17, max_iter=1000)
```

```python
lr_grid = GridSearchCV(lr, lr_params, cv=5)
```

```python
model_lr=lr_grid.fit(x_train, y_train)
```

```python
model_lr.best_params_
```

```python
predict_lr = model_lr.predict(x_test)
```

```python
print(f'LogisticRegression Accuracy: { accuracy_score(y_test, predict_lr)}')
```

```python
knn_params = {'n_neighbors': np.arange(3, 7, 1)}   # KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn_grid = GridSearchCV(knn, knn_params, cv=5)
```

```python
model_knn=knn_grid.fit(x_train, y_train)
```

```python
model_knn.best_params_
```

```python
predict_knn = model_knn.predict(x_test)
```

```python
print(f'KNeighborsClassifier Accuracy: { accuracy_score(y_test, predict_knn)}')
```

# 4. Сравнение качества моделей 

Лучшей моделью в соотвествии метрикой качества Accuracy является GradientBoostingClassifier

С помощью метода predict_proba получим вероятности принадлежности объектов тестовой выборки к классам. Построим график roc_auc для тестовой выборки и выведим значение auc.

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
probas = model_knn.fit(x_train, y_train).predict_proba(x_test) # KNeighborsClassifier
fpr, tpr, _ = roc_curve(y_test, probas[:, 1])
auc = roc_auc_score(y_test, probas[:, 1])
plt.plot(fpr, tpr, label="auc=" + str(auc))
plt.legend(loc=4)
```

```python
probas = model_lr.fit(x_train, y_train).predict_proba(x_test) # LogisticRegression
fpr, tpr, _ = roc_curve(y_test, probas[:, 1])
auc = roc_auc_score(y_test, probas[:, 1])
plt.plot(fpr, tpr, label="auc=" + str(auc))
plt.legend(loc=4)
```

```python
probas = model_gbc.fit(x_train, y_train).predict_proba(x_test) # GradientBoostingClassifier
fpr, tpr, _ = roc_curve(y_test, probas[:, 1])
auc = roc_auc_score(y_test, probas[:, 1])
plt.plot(fpr, tpr, label="auc=" + str(auc))
plt.legend(loc=4)
```

```python
probas = model_rfc.fit(x_train, y_train).predict_proba(x_test) # RandomForestClassifier
fpr, tpr, _ = roc_curve(y_test, probas[:, 1])
auc = roc_auc_score(y_test, probas[:, 1])
plt.plot(fpr, tpr, label="auc=" + str(auc))
plt.legend(loc=4)
```

```python
probas = model_svc.fit(x_train, y_train).predict_proba(x_test) # SVC
fpr, tpr, _ = roc_curve(y_test, probas[:, 1])
auc = roc_auc_score(y_test, probas[:, 1])
plt.plot(fpr, tpr, label="auc=" + str(auc))
plt.legend(loc=4)
```

После оценки точности каждого алгоритма, получаем, что лучше всех зарекомендовал алгоритм GradientBoostingClassifier (Градиентный бустинг)
